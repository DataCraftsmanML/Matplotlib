```python
x = np.array([1,10])
y = np.array([1,10])

plt.plot(x, y)

plt.title("Sports Watch Data")
plt.xlabel("Average Pulse")
plt.ylabel("Calorie Burnage")

plt.show()
```

fontdict is a parameter in Matplotlib that allows you to customize text properties (like font family, size, and color) in plots.

```python
x = np.array([1,10])
y = np.array([1,10])

font1 = {'family':'serif','color':'blue','size':20}
font2 = {'family':'serif','color':'darkred','size':15}

plt.title("Sports Watch Data", fontdict = font1)
plt.xlabel("Average Pulse", fontdict = font2)
plt.ylabel("Calorie Burnage", fontdict = font2)

plt.plot(x, y)
plt.show()
```

You can add a title to the entire figure with the suptitle() function:


```python
#plot 1:
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(1, 2, 1)
plt.plot(x,y)
plt.title("SALES")

#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(1, 2, 2)
plt.plot(x,y)
plt.title("INCOME")

plt.suptitle("MY SHOP")
plt.show()
```

Matplotlib Scatter

*With Pyplot, you can use the scatter() function to draw a scatter plot.
*The scatter() function plots one dot for each observation. It needs two arrays of the same length, one for the values of the x-axis, and one for values on the y-axis:

*Visualizing Relationships: Shows how two variables relate to each other.

```python
x = np.array([0,1,2,3])
y = np.array([4,7,9,3])
plt.scatter(x,y)

x = np.array([1,2,3,1])
y = np.array([5,8,0,2])
plt.scatter(x,y)

plt.show()
```

How to Use the ColorMap

cmap stands for "color map," which is a way to represent data values with colors. 'Viridis' is a specific colormap that transitions from dark blue to yellow, designed for clear and uniform data visualization.

```python
x = np.array([0,1,2,3])
y = np.array([4,7,9,3])
colour = np.array([100,60,40,0])
plt.scatter(x,y,c=colour, cmap='viridis')
plt.show()
```

```python
x = np.array([0,1,2,3])
y = np.array([4,7,9,3])
colour = np.array([100,60,40,0])
plt.scatter(x,y,c=colour, cmap='viridis')

plt.colorbar()
plt.show()
```

The random part is just the module nameit means "randomly generated."

100: Defines the range of random integers (0 to 99).
size=100: Specifies that you want 100 integers in the output array.


```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 15 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```

Matplotlib Pie Charts

Creating Pie Charts
With Pyplot, you can use the pie() function to draw pie charts:

```python
y = np.array([25,50,75,100])
plt.pie(y)
plt.show() 
```

Labels

```python
y = np.array([25,50,75,100])
Labels = np.array(["jia","meet","asha","bhavin"])
plt.pie(y,labels=Labels)
plt.show() 
```

```python
y = np.array([25,50,75,100])
Labels = np.array(["jia","meet","asha","bhavin"])
plt.pie(y, labels=Labels, startangle = 150)
plt.show() 
```

Explode

```python
y = np.array([25,50,75,100])
Labels = np.array(["jia","meet","asha","bhavin"])
myexplode = [0.1, 0.2, 0.3, 0.4]
plt.pie(y, labels=Labels, explode=myexplode)
plt.show() 
```

```python
Shadow

'r' - Red
'g' - Green
'b' - Blue
'c' - Cyan
'm' - Magenta
'y' - Yellow
'k' - Black
'w' - White
```

```python
y = np.array([25,50,75,100])
Labels = np.array(["jia","meet","asha","bhavin"])
mycolors = ["black", "hotpink", "b", "#4CAF50"]
plt.pie(y, labels=Labels, colors = mycolors)
plt.show() 
```

Legend
To add a list of explanation for each wedge, use the legend() function:

```python
y = np.array([25,50,75,100])
Labels = np.array(["jia","meet","asha","bhavin"])
plt.pie(y, labels=Labels, colors = mycolors)
plt.legend(title = "Four Fruits:")
plt.show() 
```
